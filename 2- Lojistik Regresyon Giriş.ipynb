- Lojistik regresyonda amacımız, sınıflandırma problemi için bağımlı ve bağımsız değişkenler arasındaki ilişkiyi tanımlayan doğrusal bir model kurmaktır.
- Lojistik regresyon bir regresyon değildir. Sınıflandırmada kullanılır. Eşik değerin altındakileri reddederken üstündekileri kabul eder.
- Bağımlı değişken kategoriktir.
- Adını bağımlı değişkene uygulanan "logit dönüşüm"den alır.
- Doğrusal regresyonda aranan varsayımlar burada aranmadığı için daha esnek kullanılabilirliği vardır.
- Bağımlı değişkenin 1 olarak tanımlanan değerinin gerçekleşme olasılığı hesaplanır. Dolayısıyla bağımlı değişkenin alacağı değer ile ilgilenmez.
- Lojistik fonks. sayesinde üretilen değerler 0-1 arasında değerler olur. 
<br></br>

- ***Lojistik Dağılım Fonks:*** Bağımsız değişken değerleri yerine girildiğinde bağımlı değişkenin 1 olma olasılığıdır.



<br></br>

- ***Lojistik Dağılım Fonks. Odds'u (Dönüşümü):*** Dağılım fonksiyonunu / 1-dağılım fonksiyonu




- ***Lojistik Regresyon Modeli:*** Gördüğümüz gibi Odds'un e tabanında logaritmasını aldığımızda ***doğrusal regresyon*** modelimiz çıkıyor.




- ***Peki bu ß katsayılarını nasıl bulacağız?:*** Doğrusal reg. (gerçek-tahmini)^2 değerini min yapan değerleri buluyorduk. Burada da benzer yöntemi kullanıyoruz. (gerçek-tahmin)
