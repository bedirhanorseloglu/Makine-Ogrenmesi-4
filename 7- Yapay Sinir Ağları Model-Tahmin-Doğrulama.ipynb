```python
# Kütüphaneler

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import mean_squared_error , r2_score , confusion_matrix , accuracy_score , roc_auc_score , roc_curve , classification_report
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale , StandardScaler     # standardizasoyon işlemleri
from sklearn import model_selection
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn import neighbors
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LogisticRegression


from warnings import filterwarnings
filterwarnings("ignore")     # önceki bölümlerde çalıştırdığımızda çıkan gereksiz uyarı mesajları çıkmasın diye
```

```python
df = pd.read_csv("Veri ve Fotoğraflar\diabetes.csv")
df.head()


# Bağımlı değişkenimiz "Outcome". Yani burada kişi diyabet hastası mı değil mi bunu inceliyoruz. 1, hasta ; 0, hasta değil
```

```python
y = df["Outcome"]

x = df.drop(["Outcome"] , axis=1)

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.30 , random_state=42)
```

# MODEL

```python
ysa_model = MLPClassifier().fit(x_train , y_train)
```

```python
# Modelimizdeki ağırlıkları(w) görüntüleyebiliriz.

ysa_model.coefs_
```

```python
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

?ysa_model

# Doğrusal bir regresyon için "relu" aktivasyon fonks. , sınıflandırma regresyonu için "logistic" aktivasyon fonks. kullanmalıyız.

# solver -> modelimizde kullanmış olduğumuz ağırlıkları optimize etmek için kullanılacak olan yöntem. "adam" büyük boyutlu veri setlerinde daha iyi çalışırken "lbfgs" daha küçük boyutlu veri setlerinde daha iyi çalışır.

# alpha -> ridge ve lasso'da kullandığımız ceza terimi
```

# TAHMİN

```python
y_pred = ysa_model.predict(x_test)

accuracy_score(y_test , y_pred)
```

# MODEL TUNING (DOĞRULAMA)

```python
ysa_params = {"hidden_layer_sizes": [(10,10) , (100,100,100) , (100,100) , (3,5)] , "alpha": [0.1 , 0.01 , 0.03 , 0.005 , 0.0001 , 1 ,5]}
```

```python
ysa_cv_model = GridSearchCV(ysa_model , ysa_params , cv=10 , n_jobs=-1 , verbose=2).fit(x_train ,y_train )
```

```python
ysa_cv_model.best_params_
```

```python
ysa_cv_model.best_score_
```

```python
# Final modelimiz

ysa_tuned = MLPClassifier(alpha= 5 , hidden_layer_sizes= (100,100,100)).fit(x_train , y_train)
```

```python
y_pred = ysa_tuned.predict(x_test)

accuracy_score(y_test , y_pred)
```
